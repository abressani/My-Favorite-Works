```python
# From Lab 7 

# G. Besla
# with code from R. Hoffman, R. Li and E. Patel

# import modules
import numpy as np
import astropy.units as u
from astropy.constants import G

# import plotting modules
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import matplotlib.patches as patches

# my modules
from ReadFile import Read
from CenterOfMass2 import CenterOfMass

R200 = 278.737
```

```python
# Create a COM of object for Merger Remnant halo (particle type=1) Using Code from Homework 4
COMH = CenterOfMass("Merger_801.txt",1)
```

```python
# Compute COM of Remnant using halo particles
COMP = COMH.COM_P(0.1, 2)
```

```python
# Determine positions of halo particles relative to COM 
xD = COMH.x - COMP[0].value 
yD = COMH.y - COMP[1].value 
zD = COMH.z - COMP[2].value 
```

```python
# Plotting
# Merger Halo Density 
fig, ax= plt.subplots(figsize=(12, 10))

plt.hist2d(xD, yD, bins=2500, norm=LogNorm(), cmap='inferno')
plt.style.use('dark_background')

cbar = plt.colorbar()
cbar.set_label("Number of disk particle per bin", fontsize=15)

# Adding R200 circle
circle = patches.Circle((0, 0), R200, edgecolor='white', facecolor='none', label='R200')
ax.add_patch(circle)

# Add axis labels
plt.xlabel('x (kpc) ', fontsize=22)
plt.ylabel('y (kpc) ', fontsize=22)
plt.title("Remnant Halo", fontsize=22)

# Set axis limits
plt.ylim(-1200,1200)
plt.xlim(-1200,1200)
plt.legend()

# Adjust tick label font size
label_size = 22
matplotlib.rcParams['xtick.labelsize'] = label_size 
matplotlib.rcParams['ytick.labelsize'] = label_size
```
